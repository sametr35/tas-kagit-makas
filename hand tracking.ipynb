```python
import mediapipe as mp
import cv2
import time
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
cap = cv2.VideoCapture(0)
user1 = (0,255,255)
user2 = (0,255,255)
point1 = int(100)
point2 = int(100)
a=0
with mp_hands.Hands(min_detection_confidence=0.1, min_tracking_confidence=0.1) as hands: 
    start_time = time.time()
    
    a1_length = 79
    a2_length = 270
    a3_length = 579
    a4_length = 370
    while cap.isOpened():
        ret, frame = cap.read()
        
        #width  = cap.get(cv2.CAP_PROP_FRAME_WIDTH)   
        #height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        # BGR 2 RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip on horizontal
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Detections
        results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        
        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        cv2.rectangle(image,(330,60),(630,420),user1,2)
        cv2.rectangle(image,(310,60),(10,420),user2,2)
        cv2.line(image, (a1_length,40), (a2_length,40), (0,0,255), 6)
        cv2.line(image, (a3_length,40), (a4_length,40), (0,0,255), 6)
        cv2.putText(image, str(point1), (50,30), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,0,0) , 2)
        cv2.putText(image,str(point2), (350,30), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,0,0) , 2)
        # Rendering results
        if results.multi_hand_landmarks:
            for hand in results.multi_hand_landmarks:
                
                x,y = hand.landmark[12].x*640 ,hand.landmark[12].y*480
                x1,y1 = hand.landmark[8].x*640 ,hand.landmark[8].y*480
                x2,y2 = hand.landmark[16].x*640 ,hand.landmark[16].y*480
                x3,y3 = hand.landmark[20].x*640 ,hand.landmark[20].y*480
                x4,y4 = hand.landmark[4].x*640 ,hand.landmark[4].y*480
                x5,y5 = hand.landmark[0].x*640 ,hand.landmark[0].y*480
                
                
                
                
                if(330<x<630 and 60<y<420):
                    if(x<x2 and x1<x2 and abs(x-x5)>150 and abs(x1-x5)>150):
                        cv2.putText(image, "makas", (int(x),int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(0,255,0) , 2)
                        sag = "makas"
                    elif(x<x4 and x1<x4 and x2<x4 and x3<x4):
                        cv2.putText(image, "kagit", (int(x),int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(0,255,0) , 2)
                        sag="kagit"
                    elif(abs(x-x5)<200 and abs(x1-x5)<200 and abs(x2-x5)<200 and abs(x3-x5)<200):
                        cv2.putText(image, "tas", (int(x),int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(0,255,0) , 2)
                        sag="tas"
                elif(10<x1<310 and 60<y1<420):
                    if(x>x2 and x1>x2 and abs(x-x5)>150 and abs(x1-x5)>150):
                        cv2.putText(image, "makas", (int(x),int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(0,255,0) , 2)
                        sol="makas"
                    elif(x>x4 and x1>x4 and x2>x4 and x3>x4):
                        cv2.putText(image, "kagit", (int(x),int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(0,255,0) , 2)
                        sol="kagit"
                    elif(abs(x-x5)<200 and abs(x1-x5)<200 and abs(x2-x5)<200 and abs(x3-x5)<200):
                        cv2.putText(image, "tas", (int(x),int(y)), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(0,255,0) , 2)
                        sol="tas"
                
            end_time = (time.time()-start_time)%6
            
            if float(4)<end_time<float(5) and a==0:
                cv2.putText(image, "1", (300,200), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,255,255) , 2)
            elif float(3)<end_time<float(4)  and a==0:
                cv2.putText(image, "2", (300,200), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,255,255) , 2)
            elif float(2)<end_time<float(3)  and a==0:
                cv2.putText(image, "3", (300,200), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,255,255) , 2)
            elif float(1)<end_time<float(2)  and a==0:
                cv2.putText(image, "4", (300,200), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,255,255) , 2)
            elif float(0)<end_time<float(1) and a==0:
                cv2.putText(image, "5", (300,200), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,255,255) , 2)
            elif(float(5)<end_time<float(6) and a==0) and a==0:
                cv2.putText(image, "0", (300,200), cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,255,255) , 2)
                start_time = time.time()-5000
                if (sag=="makas" and sol =="makas"):
                    cv2.putText(image, "draw", (250,200), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,215,255) , 4)
                    user1 = (0,255,255)
                    user2 = (0,255,255)
                elif(sag=="makas" and sol =="kagit"):
                    #cv2.putText(image, "won", (70,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    #cv2.putText(image, "lost", (450,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user1 = (0,255,0)
                    user2 = (0,0,255)
                    a4_length = a4_length+40
                    point1=point1-20
                elif(sag=="makas" and sol =="tas"):
                    #cv2.putText(image, "won", (450,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    #cv2.putText(image, "lost", (70,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user2 = (0,255,0)
                    user1 = (0,0,255)
                    a4_length = a4_length+40
                    point2=point2-20
                    
                elif(sag=="kagit" and sol =="kagit"):
                    user1 = (0,255,255)
                    user2 = (0,255,255)
                    #cv2.putText(image, "draw", (250,200), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                elif(sag=="kagit" and sol =="makas"):
                    #cv2.putText(image, "won", (70,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    #cv2.putText(image, "lost", (450,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user2 = (0,255,0)
                    user1 = (0,0,255)
                    a4_length = a4_length+40
                    point2=point2-20
                elif(sag=="kagit" and sol =="tas"):
                    #cv2.putText(image, "won", (450,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    #cv2.putText(image, "lost", (70,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user1 = (0,255,0)
                    user2 = (0,0,255)
                    a2_length = a2_length-40
                    point1=point1-20
                elif(sag=="tas" and sol =="kagit"):
                    #cv2.putText(image, "won", (450,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                   # cv2.putText(image, "lost", (70,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user2 = (0,255,0)
                    user1 = (0,0,255)
                    a4_length = a4_length+40
                    point2=point2-20
                elif(sag=="tas" and sol =="tas"):  
                    cv2.putText(image, "draw", (250,200), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user1 = (0,255,255)
                    user2 = (0,255,255)
                elif(sag=="tas" and sol =="makas"):
                    #cv2.putText(image, "won", (70,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    #cv2.putText(image, "lost", (450,100), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
                    user1 = (0,255,0)
                    user2 = (0,0,255)
                    a2_length = a2_length-40
                    point1=point1-20
        if(int(a2_length-a1_length)<10):
            cv2.putText(image, "player2 won", (150,200), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
            a=1
            a2_length=-10
            a1_length=-10
        elif(int(a3_length-a4_length)<10):
            cv2.putText(image, "player1 won", (150,200), cv2.FONT_HERSHEY_SIMPLEX, 2,(0,215,255) , 4)
            a=1
            a3_length=-10
            a4_length=-10
                
                    
              
          
            
        cv2.imshow('Hand Tracking', image)


        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
      
            

cap.release()
cv2.destroyAllWindows()
```
